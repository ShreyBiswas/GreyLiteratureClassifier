```python
import os,json
from tqdm.notebook import tqdm

import pandas as pd
```

```python
path = '../../data/labelled/'

data = []

for subdir in os.listdir(path):
    if os.path.isdir(os.path.join(path, subdir)):
        print(subdir)
        data.append(pd.read_json(os.path.join(path, subdir,f'{subdir}.json'),orient='records'))
```

```python

data = pd.concat(data)

data.head()
```

```python
# filter down and drop unknown classes
write_data = data[['url','text','relevance','multiclasses']].dropna(subset=['multiclasses'])
write_data.head()
```

```python
write_data.to_json('../../data/labelled/relevant.json',orient='records')
```

```python
def merge_data():
    relevant = pd.read_json('../../data/labelled/relevant.json',orient='records')
    irrelevant = pd.read_json('../../data/labelled/irrelevant.json',orient='records')
    data = pd.concat([relevant,irrelevant])
    return data

merged_data = merge_data()
print(merged_data.info())
merged_data.head()
```

```python

merged_data.to_json('../../data/labelled/data.json',orient='records')
```
