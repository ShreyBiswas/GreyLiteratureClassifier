```python
import os,json
from tqdm.notebook import tqdm

import pandas as pd
```

```python
path = 'data/labelled'

data = {}

for subdir in os.listdir(path):
    if os.path.isdir(os.path.join(path, subdir)):
        print(subdir)
        for file in tqdm(os.listdir(os.path.join(path, subdir))):
            if file.endswith('.json'):
                with open(os.path.join(path, subdir, file), 'r') as f:
                    data[file] = json.load(f)

data = pd.DataFrame.from_dict(data,orient='index')
data.head()
```

```python
# get null classes
x = data[data['class'].isnull()]
```

```python
# filter down and drop unknown classes
write_data = data[['url','text','class']].dropna(subset=['class'])
write_data.head()
```

```python
write_data.to_json('data/labelled/relevant.json',orient='records')
```

```python
def merge_data():
    relevant = pd.read_json('data/labelled/relevant.json',orient='records')
    irrelevant = pd.read_json('data/labelled/irrelevant.json',orient='records')
    data = pd.concat([relevant,irrelevant])
    return data

merged_data = merge_data()
print(merged_data.info())
merged_data.head()
```

```python

merged_data.to_json('data/labelled/data.json',orient='records')
```
